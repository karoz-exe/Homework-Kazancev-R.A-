```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Подзадание №1
import os
import pandas as pd
print(f"Текущая директория: {os.getcwd()}")
df = pd.read_csv("/content/data titanic.csv")
```

```python
# Подзадание №2
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.countplot(x='Survived', data=df)
plt.title('Распределение выживших')
plt.xlabel('Выжил (1 - да, 0 - нет)')
plt.ylabel('Количество пассажиров')

plt.subplot(2, 2, 2)
sns.countplot(x='Pclass', data=df)
plt.title('Распределение по классам')
plt.xlabel('Класс')
plt.ylabel('Количество пассажиров')

plt.subplot(2, 2, 3)
sns.histplot(x='Age', data=df, kde=True) #kde=True добавляет кривую плотности
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')

plt.subplot(2, 2, 4)
sns.countplot(x='Sex', data=df)
plt.title('Распределение по полу')
plt.xlabel('Пол')
plt.ylabel('Количество пассажиров')

plt.tight_layout()
plt.show()
```

```python
# Подзадание №3
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=df)
plt.title('Boxplot возраста')
plt.xlabel('Возраст')
plt.show()
```

```python
# Подзадание №4
print("Интерпретация boxplot'а:\n")
print("Медиана возраста пассажиров составляет около 28 лет. ")
print("Есть значительное количество выбросов (значения за пределами усов) в данных о возрасте.")
print("Разброс значений возраста довольно велик, что видно по высоте boxplot'а.")
```

```python
# Подзадание №5
plt.subplot(1, 2, 1)
df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['Не выжил', 'Выжил'], startangle=90)
plt.title('Доля выживших')

plt.subplot(1, 2, 2)
df['Pclass'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['1 класс', '3 класс', '2 класс'], startangle=90)
plt.title('Доля пассажиров по классам')

plt.tight_layout()
plt.show()
```

```python
# Подзадание №6
numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
sns.pairplot(df[numeric_cols])
plt.show()
```

```python
# Подзадание №7
data_sunburst = df.groupby(['Pclass', 'Sex'])['PassengerId'].count().reset_index()
fig_sunburst = px.sunburst(data_sunburst, path=['Pclass', 'Sex'], values='PassengerId',
             title='Количество пассажиров по классу и полу')
fig_sunburst.show()
```
