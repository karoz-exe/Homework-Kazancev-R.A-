**Практическое задание № 7**

Собрать данные с сайта lifehacker.ru из рубрики Технологии с помощью библиотек requests и BeautifulSoup.

Подзадачи:
* Определить формат ссылки для пагинации - 1 балл
* В разметке страницы найти уникальные классы/идентификаторы блока с названием и содержанием материала- 3 балла
* Получить содержимое десяти страниц списка материалов, выделить из него ссылки на каждый материал - 2 балла
* Перебрать все полученные ссылки и получить html-код каждого материала - 1 балл
* Распарсить полученный текст разметки с помощью BeautifulSoup, вытащив по каждой ссылке заголовок и содержание материала - 2 балл
* Создать датайфрейм с полученными данными - 1 балл

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

article_container_selector = "div.article-card__small-wrapper" # Контейнер для каждой статьи
article_title_selector = "a.lh-small-article-card__title"  # Заголовок статьи
article_url_attribute = "href = /kakoi-budet-skladnoi-iphone/"  # Атрибут ссылки на статью
article_text_selector = "lh-small-article-card__link" # Текст статьи

base_url = "https://lifehacker.ru/topics/technology/?page=1"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

all_titles = []
all_texts = []

for page_number in range(1, 11):
    url = base_url + str(page_number) + "/" if page_number > 1 else base_url + "/"
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Проверка HTTP-кодов ошибок
        print(f"Загружена страница: {url}")
        time.sleep(1)  # Задержка, чтобы не перегружать сервер
        soup = BeautifulSoup(response.content, 'html.parser')

        articles = soup.select(article_container_selector)
        print(f"Найдено статей на странице {page_number}: {len(articles)}")

        for article in articles:
            title_tag = article.select_one(article_title_selector)
            if title_tag:
                title = title_tag.text.strip()
                article_url = title_tag[article_url_attribute]

                try:
                    article_response = requests.get(article_url, headers=headers)
                    article_response.raise_for_status()
                    article_soup = BeautifulSoup(article_response.content, 'html.parser')
                    text_tag = article_soup.select_one(article_text_selector)
                    if text_tag:
                        text = text_tag.get_text(" ", strip=True)
                        all_titles.append(title)
                        all_texts.append(text)
                        print(f"Обработана статья: {title}")
                    else:
                        print(f"Текст статьи не найден на странице: {article_url}")
                except requests.exceptions.HTTPError as e:
                    print(f"Ошибка при загрузке статьи {article_url}: {e}")
                except Exception as e:
                    print(f"Ошибка при обработке статьи {article_url}: {e}")
            else:
                print(f"Заголовок статьи не найден на странице: {url}")

    except requests.exceptions.HTTPError as e:
        print(f"Ошибка при загрузке страницы {url}: {e}")
        break  # Прекращаем цикл, если произошла ошибка
    except Exception as e:
        print(f"Произошла неизвестная ошибка: {e}")


df = pd.DataFrame({'title': all_titles, 'text': all_texts})
print(df)
```
