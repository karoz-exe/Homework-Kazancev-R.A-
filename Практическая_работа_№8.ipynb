hint: используйте методы conn.cursor(), cursor.execute(), cursor.fetchall(). В SQL-запросе не забудьте использовать LIMIT 10 и указать нужные вам столбцы

```python
import requests
import pandas as pd
import psycopg2
#API OF ICE AND FIRE
# Информации обо всех книгах
books_url = "https://www.anapioficeandfire.com/api/books"
response_books = requests.get(books_url)
books_data = response_books.json()
books_df = pd.DataFrame(books_data)
# Информации обо всех домах Вестероса
houses_url = "https://www.anapioficeandfire.com/api/books"
response_houses = requests.get(houses_url)
houses_data = response_houses.json()
houses_df = pd.DataFrame(houses_data)
# Информации о домах Вестероса с девизом
houses_with_titles_url = "https://www.anapioficeandfire.com/api/books"
response_houses_titles = requests.get(houses_with_titles_url)
houses_titles_data = response_houses_titles.json()
houses_titles_df = pd.DataFrame(houses_titles_data)

hostname = "hh-pgsql-public.ebi.ac.uk"  # hostname
database = "pfmegrnargs"  # Имя базы данных
user = "reader"  # Имя пользователя
password = "NWDMCE5xdipIjRrp"  # Пароль
try:
    # Подключение к БД
    conn = psycopg2.connect(host=hostname, database=database, user=user, password=password)
    cur = conn.cursor()

    # Получение 10 строк из таблицы rnc_database
    cur.execute("SELECT * FROM rnc_database LIMIT 10;")
    rows = cur.fetchall()
    columns = [desc[0] for desc in cur.description]
    rnc_df = pd.DataFrame(rows, columns=columns)

    # Получение значений указанных столбцов для 10 строк
    cur.execute("SELECT display_name, num_sequences, num_organisms, url FROM rnc_database LIMIT 10;")
    rows_selected = cur.fetchall()
    rnc_selected_df = pd.DataFrame(rows_selected, columns=['display_name', 'num_sequences', 'num_organisms', 'url'])


    # Закрытие курсора и соединения
    cur.close()
    conn.close()

except psycopg2.Error as e:
    print(f"Ошибка при работе с БД: {e}")

# Вывод DataFrame (для проверки)
print("Books DataFrame:\n", books_df.head())
print("\nHouses DataFrame:\n", houses_df.head())
print("\nHouses with Titles DataFrame:\n", houses_titles_df.head())
print("\nRNC Database DataFrame:\n", rnc_df.head())
print("\nRNC Database Selected Columns DataFrame:\n", rnc_selected_df.head())
```
